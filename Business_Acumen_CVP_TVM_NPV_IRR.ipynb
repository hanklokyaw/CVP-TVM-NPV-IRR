```python
### Professer Larry CMP-TMV-NPV-IRR

import math
from IPython import display
import time

def PV(fv, rate, nper):
    pv = fv / (1 + rate) ** nper
    return pv

def FV(pv, rate, nper):
    fv = pv * (1 + rate) ** nper
    return fv

def RATE(pv, fv, nper):
    rate = (math.pow((fv/pv),1/nper)) - 1
    return rate

def NPER(pv, fv, nper):
    nper = math.log(fv/pv, 1+rate)
    return nper

calculation_continue = True

while calculation_continue:

    choice = input("What do you want to calculate?\n "
                       "1.Present Value\n "
                       "2.Future Value\n "
                       "3.Interest Rate\n "
                       "4.Year(Payment times)\n\n\n"
                       "Enter 'Exit' to quit the app.\n")
    if 'exit' in choice:
        time.sleep(1)
        display.clear_output()
        calculation_continue = False
    
    elif choice == '1':
        time.sleep(1)
        display.clear_output()
        fv = float(input("What is your Future Value: "))
        rate = float(input("What is your Interest Rate (such as 0.01 or 0.5): "))
        nper = float(input("What is your Year: "))
        pv = PV(fv, rate, nper)
        print(">>> You present value is: $",round(pv))

    elif choice == '2':
        time.sleep(1)
        display.clear_output()
        pv = float(input("What is your Present Value: "))
        rate = float(input("What is your Interest Rate (such as 0.01 or 0.5): "))
        nper = float(input("What is your Year: "))
        fv = FV(pv, rate, nper)
        print(">>> Your future value is: $",fv)

    elif choice == '3':
        time.sleep(1)
        display.clear_output()
        pv = float(input("What is your Present Value: "))
        fv = float(input("What is your Future Value: "))
        nper = float(input("What is your Year: "))
        rate = RATE(pv, fv, nper)
        print(">>> Your interest rate is: ",int(RATE *100),"%")

    elif choice == '4':
        time.sleep(1)
        display.clear_output()
        pv = float(input("What is your Present Value: "))
        fv = float(input("What is your Future Value: "))
        rate = float(input("What is your Interest Rate (such as 0.01 or 0.5): "))
        nper = NPER(pv, fv, rate)
        print("Number of year is: ",nper," years")

    else:
        time.sleep(1)
        display.clear_output()
        print("Sorry, wrong choice. Try again!")
```
